```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\soura\Desktop\Iris.csv")
```

```python
df
```

```python
df['Species'].unique()
```

```python
df['Species'].replace({'Iris-setosa':'1', 'Iris-versicolor':'2', 'Iris-virginica':'3'},inplace=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']],df['Species'],test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn import linear_model
```

```python
lr=linear_model.LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.predict(x_test)
```

```python
x_test   #{'Iris-setosa', 'Iris-versicolor', 'Iris-virginica'}
```

```python
lr.score(x_test,y_test)
```
