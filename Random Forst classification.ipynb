```python
from sklearn import datasets
```

```python
df=datasets.load_iris()
```

```python
df
```

```python
print(df.target_names)
print(df.feature_names)
```

```python
import pandas as pd
data=pd.DataFrame({'sepal length':df.data[:,0],
             'sepal width':df.data[:,1],
             'petal length':df.data[:,2],
             'petal width':df.data[:,3],
             'species':df.target})
```

```python
data.head()
```

```python
x=data[['sepal length', 'sepal width', 'petal length', 'petal width']]
y=data['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf=RandomForestClassifier(n_estimators=100,criterion='gini')  # n_estimators means number of tree is to be included
```

```python
clf.fit(x_train,y_train)
```

```python
clf.predict(x_test)
```

```python
x_test  # (0: setosa, 1: versicolor, 2: virginica)
```

```python
clf.score(x_test,y_test)
```
